```python
import glob
import os
A=glob.glob("*.png")
#print(A)
N=len(A)
for j in range(1,N+1):
    src="t:"+str(j)+"-150 - .png"
    if src in A:
        b=str(j)
        if len(b)==1:
            z='00'+b
        if len(b)==2:
            z='0'+b
        if len(b)==3:
            z=b
        bx='img'+z+'.png'
        print(src,bx)
        os.rename(src,bx)
```
